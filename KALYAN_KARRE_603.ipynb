1. Write Python code and use MapReduct to count occurrences of each word in the first text file (file.txt). How many times each word is repeated?

```python
from collections import defaultdict
from functools import reduce
from spellchecker import SpellChecker
```

```python
# Read the file into a list of words
with open('file1.txt', 'r', encoding='utf-8') as f:
    words = f.read().split()
```

```python
# Map each word to a tuple (word, 1) and group by word
word_counts = defaultdict(int)
for word in words:
    word_counts[word] += 1
```

```python
# Reduce by summing the values for each word
total_counts = reduce(lambda x, y: x + y, word_counts.values())
```

```python
# Print the results
for word, count in sorted(word_counts.items(), key=lambda x: x[1], reverse=True):
    print(f'{word}: {count} ({count/total_counts*100:.2f}%)')
```

2. From the second text file (file2.txt), write Python code and use MapReduct to count how many times non-English words (names, places, spells etc.) were used. List those words and how many times each was repeated.

```python
# Create an english dictionary 
english_dict = SpellChecker(language='en')
#english_dict = SpellChecker(language='en')
```

```python
# Read the file into a list of words
with open('file2.txt', 'r', encoding='utf-8') as f:
    words = f.read().split()
```

```python
# Map each non-English word to a tuple (word, 1) and group by word
non_english_counts = defaultdict(int)
for word in words:
    if word.lower() not in english_dict:
        non_english_counts[word] += 1
```

```python
# Reduce by summing the values for each non-English word
total_counts = reduce(lambda x, y: x + y, non_english_counts.values())
```

```python
# Print the results
for word, count in sorted(non_english_counts.items(), key=lambda x: x[1], reverse=True):
    print(f'{word}: {count} ({count/total_counts*100:.2f}%)')
```
